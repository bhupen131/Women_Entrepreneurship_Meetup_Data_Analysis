```python
sentence = df_startup["About_Startup"].str.cat(sep='. ')
```

### Creating Tokens

```python
from nltk.tokenize import RegexpTokenizer

tokenizer = RegexpTokenizer('\w+')

tokens = tokenizer.tokenize(sentence)
tokens[:8]
```

### Converting The Text To Lowercase 

```python
words = []
for word in tokens:
    words.append(word.lower())
words[:8]
```

### Creating a New List (For Words With No Stopwords)

```python
no_sw = []

for word in words:
    if word not in sw:
        no_sw.append(word)
```

```python
import re
punctuation = re.compile(r'[-.?!,:;()|0-9]')
```

```python
post_punctuation = []
for words in no_sw:
    word = punctuation.sub('',words)
    if len(word)>0:
        post_punctuation.append(word)
```

### Importing FreqDist For Determining The Frequencies Of N-grams

```python
from nltk.probability import FreqDist
```

### Unigram

```python
unigram_freq = nltk.FreqDist(post_punctuation)
unigram_freq
```

```python
unigram_freq = unigram_freq.most_common(98)
unigram_freq[:5]
```

```python
df_unigram = pd.DataFrame(unigram_freq,columns=['unigram','unigram_freq'])
df_unigram.head()
```

```python
df_unigram['unigram'] = df_unigram['unigram'].str.title()
```

### Bigram

```python
bigram = list(nltk.bigrams(post_punctuation))
bigram[:5]
```

```python
bigram_freq = nltk.FreqDist(bigram)
bigram_freq
```

```python
bigram_freq = bigram_freq.most_common(98)
bigram_freq[:5]
```

```python
df_bigram = pd.DataFrame(bigram_freq,columns=['bigram','bigram_freq'])
df_bigram.head()
```

### Joining The Bigram

```python
df_bigram['bigram'] = df_bigram['bigram'].agg(' '.join)
```

```python
df_bigram.head()
```

### Trigram

```python
trigram = list(nltk.ngrams(post_punctuation,3))
trigram[:5]
```

```python
trigram_freq = nltk.FreqDist(trigram)
trigram_freq
```

```python
trigram_freq = trigram_freq.most_common(98)
trigram_freq[:5]
```

```python
df_trigram = pd.DataFrame(trigram_freq,columns=['trigram','trigram_freq'])
df_trigram.tail()
```

### Joining The Trigram

```python
df_trigram['trigram'] = df_trigram['trigram'].agg(' '.join)
```

```python
df_trigram.head()
```

### Fourgram

```python
fourgram = list(nltk.ngrams(post_punctuation,4))
fourgram[:5]
```

```python
fourgram_freq = nltk.FreqDist(fourgram)
fourgram_freq
```

```python
fourgram_freq = fourgram_freq.most_common(98)
fourgram_freq[:5]
```

```python
df_fourgram = pd.DataFrame(fourgram_freq,columns=['fourgram','fourgram_freq'])
df_fourgram.tail()
```

### Joining The Fourgram

```python
df_fourgram['fourgram'] = df_fourgram['fourgram'].agg(' '.join)
```

```python
df_fourgram.head()
```

### Fivegram

```python
fivegram = list(nltk.ngrams(post_punctuation,5))
fivegram[:5]
```

```python
fivegram_freq = nltk.FreqDist(fivegram)
fivegram_freq
```

```python
fivegram_freq = fivegram_freq.most_common(98)
fivegram_freq[:5]
```

```python
df_fivegram = pd.DataFrame(fivegram_freq,columns=['fivegram','fivegram_freq'])
df_fivegram.head()
```

### Joining The Fivegram

```python
df_fivegram['fivegram'] = df_fivegram['fivegram'].agg(' '.join)
```

```python
df_fivegram.head()
```

### Concatenating All The Above Created N-grams To The Main Dataframe

## Sentiments Analysis Based On The Description About Startup

```python
from textblob import TextBlob
df_startup['sentiment'] = df_startup['About_Startup'].apply(lambda About_Startup: TextBlob(About_Startup).sentiment)
```

### Creating a New Dataframe For Sentiment

```python
df_sentiment = pd.DataFrame(df_startup['sentiment'].values.tolist(), index=df_startup.index)
df_sentiment
```

**Polarity is float which lies in the range of [-1,1] where 1 means most positive statement and -1 means a most negative statement.
<br>
<br>
Subjectivity is also a float which lies in the range of [0,1].The higher subjectivity score means that the text contains personal opinion rather than factual information and indicates that its nore subjective and whereas the lower the subjectivity score means that its more objective**

### Concatenating The 'Sentiment' Dataframe To The Main Dataframe

```python
df_startup = pd.concat([df_startup, df_sentiment], axis=1, sort=False)
```

```python
df_startup.head()
```
